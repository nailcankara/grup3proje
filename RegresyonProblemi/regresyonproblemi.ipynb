####PART2 - Machine Learning


````
df = pd.read_csv("carsclean.csv")                     #verimizi import ettik
df.drop(labels=['city','fuel'],axis=1,inplace=True)   #city  ve fuel değişkenleri işimize yaramadığı için yeniden sildik
df.head()  #df'imizin güncel hali
````

````
# df.iloc satır ve sütun seçme komutumuzdur. [:, ] virgülden önceki iki nokta tüm satırları seçmemizi sağlar.
# Virgülden sonrası sütun seçme operasyonudur. X için son sütün hariç tüm satır ve sütünları seçtik.
# y içinde tüm satırları ve son sütunu seçtik. Sonuna .values ekleyerek değerlerimizi numpy-array olarak aldık.
# Numpy-array'e dönüştürmemizin nedeni sklearn kütüphanesinin numpy verileri istemesinden dolayı.
# İsterseniz farkı görmek için "df.iloc[:, :-1]" ve "df.iloc[:, :-1].values" değerlerini ayrı ayrı çalıştırın.
X = df.iloc[:, :-1].values 
y = df.iloc[:, -1:].values
````

````
X[:1] #X'in ilk satırına bakalım
````

````
from sklearn.preprocessing import LabelEncoder   #Label encoder kategorik verileri , nümerik verilere çevirir.
labelencoder = LabelEncoder()    #LabelEncoder'ı labelencoder değişkenine atadık.
#Şimdi her bir kategorik sütunu labelencoder'la encode'licez. Neden mi? Makine yazıları anlayamaz sayıları anlar bu yüzden :)

X[:,0] = labelencoder.fit_transform(X[:,0]) #0. sütun araba markamız olan car_marker
X[:,1] = labelencoder.fit_transform(X[:,1]) #1. sütun modelimiz 
X[:,3] = labelencoder.fit_transform(X[:,3]) #3. sütun condition değerimiz
X[:,5] = labelencoder.fit_transform(X[:,5]) #5. sütun transmission değerimiz yani araç auto mu manuel mi.
X[:,6] = labelencoder.fit_transform(X[:,6]) #6. sütun arabamızın rengi olan color sütunu
X[:,7] = labelencoder.fit_transform(X[:,7]) #7. sütun ise ödeme metodu
````

Yukarıdaki koda alternatif olarak böylede yapabilirdik

```
categorical_columns = [0,1,3,5,6,7]

def categorical_to_numerical(column_number):
  X[:,column_number] = labelencoder.fit_transform(X[:,column_number])

  categorical_to_numerical(categorical_columns[i])
```

Seçim size kalmış aynı şey :)

````
#Ödeme metodumuzun ilk 5 satırı sırasıyla : Cash or Installments,Cash Only,Installments Only,Cash Only,Cash Only	'idi.

X[:5,7:]  #Şimdi ödeme metodumuzun ilk 5 satırını neye dönüştürdüğümüzü görelim.
````

Sklearn alfabetik sıralamaya göre numaralandırır. Bu yüzden örnek olarak

* Cash Only : 0
* Cash or Installments : 1
* Installments Only : 2 'dir. 

Bir sonraki aşamada OneHotEncoding yapıcaz :)

to_catogorical() metoduyla OneHotEncoding yapmak istediğimiz sütunu seçicez. Böylece 1 sütunumuz içindeki değişken veri sayısı kadar büyücek.

Az önceki örnek için;

Cash Only || Cash or Installments || Installments Only

* 1 || 0 || 0
* 0 || 1 || 0
* 0 || 0 || 1

Olucak. Gölge değişken tuzağından kaçınmamız için 1 sütunu silmemiz gerek. Installment's Only'e ihtiyacımız yok çünkü ilk 2 sütun "0" olduğunda otomatikmen Installments Only 1 anlamına gelir. O yüzden sonlarına [:,:-1] ekleyerek oluşturduğumuz OneHotEncodinglerin son sütununu dahil etmicez.

Ve son olarak eğer kategorik değişkenimiz sadece 2 eşsiz değerden oluşuyorsa OneHotEncoding yapmamıza gerek yok çünkü sütun değerleri zaten 0 veya 1 dir.

````
from keras.utils import to_categorical #Bir derin öğrenme kütüphanesi olan keras'ın nimetlerinden faydalanalım.

car_maker = to_categorical(X[:,0])[:,:-1]   #2'den fazla kategori bulundurduğu için OneHotEncoding yaptık.
model = to_categorical(X[:,1])[:,:-1]       
year = X[:,2:3].astype('float32')           #Year değişkenini 2. sütundan 3'e kadar seçtik 3 dahil değil.
condition =  X[:,3:4].astype('float32')     #Sonuna float32 ekleyerek data türümüzü garantileyelim
kilometers = X[:,4:5].astype('float32') 
transmission = X[:,5:6].astype('float32')
color = to_categorical(X[:,6])[:,:-1]       
pay_method = to_categorical(X[:,7])[:,:-1]
````

Bakalım ne el ettik ilk 5 satırı hepsinde görünteleyelim

````
def ne_elde_ettik(sutun,name):
  print(name)
  print(sutun[:3] , sutun.shape)
  print("------------------------------")

xd = [car_maker,model,year,condition,kilometers,transmission,color,pay_method]
xdxd = ["car_maker","model","year","condition","kilometers","transmission","color","pay_method"]

for i in range(len(xd)):
  ne_elde_ettik(xd[i],xdxd[i])
````

````
#Numpy'ın concantenate metoduyla tüm oluşturmuş olduğumuz(En son sütunu çıkarılmış sütünları) yanyana ekliyoruz. axis=1 sütun olarak birleştir
X = np.concatenate((car_maker,model,year,condition,kilometers,transmission,color,pay_method),axis=1) #Yeni X'imiz.
````

````
print(X , X.shape)  #560 satır 46 sütunluk bomba gibi verimiz hazır.
````

Bir sonraki bölümde makine öğrenmesi kısmına tam olarak başlıyoruz. En zevkli part :)

X bağımsız değişkenlerimiz ve y bağımlı değişkenimizi train ve test datası olarak 2 ye ayıralım. Test datamız ana verimizin %20 sini içericek. Sonuçların aynı çıkması için random_state=0 olarak kalsın.

````
from sklearn.model_selection import train_test_split   #train,test ayırmak için gerekli kütüphane
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
````

````
from sklearn.linear_model import LinearRegression    #Tekli ve Çoklu regresyon için kullandığımız kütüphane
regressor = LinearRegression()                       #LinearRegression() metodumuzu regressor değişkenine atadık.
regressor.fit(X_train, y_train)                      #.fit metoduyla X_train ve y_train'imizi modelimize uydurduk ve öğrettik.
````

Evet Bu kadar. 2 satır kod ile makinemiz öğrenmiş oldu. Şimdi gelelim sonuçlara :)

````
#.predict metodunun içine X_test'i vererek tahmin etmesini sağladık. Ve tahminlerimizi y_pred'e atadık.
y_pred = regressor.predict(X_test)
````

````
from sklearn.metrics import r2_score  #r2 score'mizi test edelim.
r2_score(y_test, y_pred)
````

Mükemmel! Tüm tahminlerimizi ve gerçek değerleri bir dataframe üzerinde görüntüleyelim
